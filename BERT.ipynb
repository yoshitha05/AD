```python
import tensorflow_hub as hub
import tensorflow_text as text
```

```python
# This is just like function, so we're taking the function where first link is for preprocessing and second link is for embeddings
preprocess_url = "https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3"
encoder_url = "https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4"
```

```python
# It does text processing i.e remove stop removal and converting into lowercase, and words, then save it into a variable called "bert_preprocess_model"
bert_preprocess_model = hub.KerasLayer(preprocess_url)
```

```python
# Here we're passing the sentence into first link i.e., preprocess_url(since this is assigned to bert_preprocess_model, we call that function)
text_test = ['Malla Reddy founded the Malla Reddy college', 'nice College indeed','I love python programming']
text_preprocessed = bert_preprocess_model(text_test)
text_preprocessed.keys()
```

```python
text_preprocessed['input_mask']
```

```python
text_preprocessed['input_type_ids']
```

```python
text_preprocessed['input_word_ids']
```

```python
bert_model = hub.KerasLayer(encoder_url)
```

```python
bert_results = bert_model(text_preprocessed)
```
